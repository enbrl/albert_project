```python
def prix(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    price=soup.find_all("span",class_="vehicle-loa-offer")
    a=re.findall("<span>(.*?)</span>",str(price))
    liste=[i.replace('\xa0', '') for i in a]
    liste=[i.replace(',99','') for i in liste]
    liste = [i for i in liste if i != '€']
    prix=[int(i) for i in liste]
    return prix
```

```python
l=[]
for i in url:
    l.append(prix(url=i))
l_price=[element for sublist in l for element in sublist]
last_price=pd.Series(l_price)
last_price
```

```python
def kil(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    kil=soup.find_all("div",class_="vehicle-zero-km")
    a=re.findall(">(.*?)</div>",str(kil))


    return a
```

```python
def names(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    name=soup.find_all("span",class_="vehicle-model")
    a=re.findall(">(.*?)</span>",str(name))
    return a
```

```python
l3=[]
for i in url:
    l3.append(names(url=i))
name=[element for sublist in l3 for element in sublist]
name=pd.Series(name)
name
```

```python
def transimission(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    moteur=soup.find_all("div",class_="vehicle-transmission")
    moteur
    a=re.findall(">\s*(.*?)\s*</span>",str(moteur))
    return a
```

```python
l4=[]
for i in url:
    l4.append(transimission(url=i))
l4=[element for sublist in l4 for element in sublist]
l4, len(l4)
```
